```python
## 数字 & 数组：
# 1、二维数组 A 中，每一行都按照从左到右递增，每一列从上到下递增。判断一个整数 key 是否在该矩阵中(耗时30min)
def find(Ai, val): 
    # 定位 val 在升序数组A 中的位置或上下边界
    if Ai is None or len(Ai) == 0 or val is None:
        return None
    sz = len(Ai)
    ibeg, iend, imid = 0, sz, sz//2
    while ibeg < sz and ibeg < iend and val != Ai[imid]:
        if val < Ai[imid]:
            iend = imid
        else:
            ibeg = imid+1
        imid = (ibeg + iend)//2
    if imid < sz:        
        if val == Ai[imid]:
            return (imid, imid) # 相等时返回下标
        elif val < Ai[imid]:
            return (imid-1, imid)
        else:
            return (imid, imid+1)
    else:
        if Ai[-1] > val:
            return (sz-1, sz)
        else:
            return (sz, sz) # 比最后一个数也大

def is_exist(A, val):
    if A is None or val is None:
        return 0
    nrow, ncol = len(A), len(A[0])
    col0 = [x[0] for x in A] # 取第一列，长度=nrow
    pos = find(col0, val)
    if pos is None:
        return 0 
    coli, colj = pos 
    if coli == colj:
        if coli < nrow:
            return 1 # 找到对应的数
        elif coli == nrow: # 看看是否在最后一行
            pos2 = find(A[-1], val)
            if pos2[0] == pos2[1] and pos2[0] < ncol:
                return 1
            else:
                return 0
    else:
        if coli == -1 and colj == 0: # 比第一个数还小
            return 0  
        if val < col0[coli]:
            # 在上一行判断看看
            pos3 = find(A[coli-1], val)
            if pos3[0] == pos3[1] and pos3[0] < ncol:
                return 1
        else:
            pos4 = find(A[coli], val)
            if pos4[0] == pos4[1] and pos4[0] < ncol:
                return 1
    return 0 

# A = [[1,2,3], [4,5,6], [4,8,9]]
# print(f'0 is in : {is_exist(A, 0)}')
# print(f'1 is in : {is_exist(A, 1)}')
# print(f'4 is in : {is_exist(A, 4)}')
# print(f'7 is in : {is_exist(A, 7)}')
# print(f'9 is in : {is_exist(A, 9)}')
# print(f'10 is in : {is_exist(A, 10)}')
      
## 2、一只青蛙一次可以跳上1级台阶，也可以跳上2级台阶。求该青蛙跳上一个 n 级的台阶总共有多少种跳法。
# 1 独立子问题划分： 青蛙跳到第k级台阶的可选路径。2 定义状态：到 第k级台阶 时的可选路径V[k]。
# 3 状态转移方程：  V[k+1] = V[k] + V[k-1]; V[0]=0; V[1]=1;V[2]=2; （注意：方程这里卡了一下，实际为斐波那契数列）
def get_ways(n):
    if n <= 0:
        return 0
    ways = [0, 1, 2]
    if n < 3:
        return ways[n]
    for i in range(3, n+1):
        ways.append(ways[i-1] + ways[i-2])
    print(f'ways={ways}')
    return ways[-1]
# print(f'ways of 2={get_ways(2)}')
# print(f'ways of 7={get_ways(7)}')

## 3. 旋转数组的最小数字：升序排序的数组经过了一次旋转 (一部分or全部都旋转回到原来的位置了)，找到其中的最小值。
# 实现框架： 二分查找处理部分旋转的情况。
## 拓展： 在旋转后的数组中查找某个值。 
# 实现框架：二分查找的边界判断调整下，区分上下界之间是否单调。
def get_roll_arr_min(arr):
    # 思路： 找逆序的位置，不断缩小范围，直到满足 v[i-] >= v[i] < v[i+1] 的值
    if arr is None or len(arr) == 0:
        return None
    sz = len(arr)
    if sz == 1:
        return arr[0]
    elif sz == 2:
        return min(arr[0], arr[1])
    elif arr[0] <= arr[sz-1] and arr[0] <= arr[1]:
        return arr[0]  # 完整旋转的情况
    beg, end, mid = 0, sz-1, sz//2
    while True: 
        if mid > 0 and arr[mid-1] > arr[mid]: # 部分旋转
            return arr[mid]
        if arr[mid] > arr[end]:
            beg = mid + 1
        else:
            end = mid 
        mid = (beg + end)//2

# print(f'res of [3,4,5,1,2]={get_roll_arr_min([3,4,5,1,2])}')
# print(f'res of [2,2,2,0,1]={get_roll_arr_min([2,2,2,0,1])}')
# print(f'res of [2,2,2,0,0]={get_roll_arr_min([2,2,2,0,0])}')
# print(f'res of [2,2,2,3,4]={get_roll_arr_min([2,2,2,3,4])}')


## 4（中等）. 求数组的最长上升子序列（不要求连续）：卡住>30min没实现
# 实现框架：引入一个辅助数组minv，存放连续长度为i的数组的最后一个数的最小值。遍历数组的元素，二分查找minv到第一个比arr[i]大的 & 替换掉（更小的maxv来找后续拓展），没有则添加到minv 增加最长子串的长度。 并设置一个数组记录第k长序列的前一个末尾值。
def get_max_subseq(nums):
    if not nums:
        return []
    
    n = len(nums)
    tail = []          # 存储最小末尾值
    tail_idx = []      # 存储对应末尾值在原数组的下标
    prev = [-1] * n    # 记录每个元素在LIS中的前驱下标    
    for i, num in enumerate(nums):
        # 二分查找：找到tail中第一个≥num的位置
        left, right = 0, len(tail)
        while left < right:
            mid = (left + right) // 2
            if tail[mid] < num:
                left = mid + 1
            else:
                right = mid
        pos = left
        
        if pos == len(tail):
            # 扩展新长度
            tail.append(num)
            tail_idx.append(i)
            if pos > 0:  # 记录前驱（非首元素时）
                prev[i] = tail_idx[pos - 1]
        else:
            # 替换当前长度的最小末尾值
            tail[pos] = num
            tail_idx[pos] = i
            if pos > 0:  # 记录前驱（非首元素时）
                prev[i] = tail_idx[pos - 1]
    
    # 回溯构造LIS具体序列
    lis = []
    idx = tail_idx[-1]  # 从LIS最后一个元素开始
    while idx != -1:
        lis.append(nums[idx])
        idx = prev[idx]  # 根据前驱回溯
    lis.reverse()        # 反转得到正序序列
    return lis

# 测试示例
# test_cases = [
#     [10, 9, 2, 5, 3, 7, 101, 18],  # 可能输出 [2, 3, 7, 18] 或 [2, 3, 7, 101]
#     [3, 5, 6, 2, 5, 4, 5],          # 可能输出 [2, 4, 5] 或 [3, 5, 6]
#     [1, 3, 5, 4, 7],                # 输出 [1, 3, 4, 7] 或 [1, 3, 5, 7]
#     [0, 8, 4, 12, 2, 10, 6, 14]     # 输出 [0, 4, 6, 14] 或 [0, 2, 6, 14]
# ]
print(get_max_subseq([10, 9, 2, 5, 3, 7, 101, 18])) # 

    
```
